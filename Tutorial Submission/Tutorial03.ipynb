```python
import numpy as np
```

```python
# 1. Create a numpy array of size 2 x 2

array01 = np.array([[1, 2], [3, 4]])
print(array01)
```

```python
# 2. Create a random numpy array of size 3 x 4

array02 = np.random.random((3, 4))
print(array02)
```

```python
# 3. Create a numpy array of ones and a numpy array of zeroes

array03 = np.ones((3, 3))
print(array03)
print()

array04 = np.zeros((3, 3))
print(array04)
```

```python
# 4. Create a random numpy array of size 6x3 and transpose it

array05 = np.random.random((6, 3))
print(array05.T)
```

```python
# 5. Create two random numpy array of size n x p and p x m and then compute the dot product

n, p, m = 3, 4, 5

array06 = np.random.random((n, p))
array07 = np.random.random((p, m))

print("First Array (n x p):")
print(array06)
print()

print("Second Array (p x m):")
print(array07)
print()

dot_product = np.dot(array06, array07)

print("Dot Product (n x m):")
print(dot_product)
```
